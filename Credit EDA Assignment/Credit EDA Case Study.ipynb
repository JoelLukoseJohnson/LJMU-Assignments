```python
#chekcing the percentage of null values in each column
(app_data.isnull().sum() * 100 / len(app_data)).round(2)
```

```python
#calculating percentage of NaN values in DataFrame
def get_perc_of_missing_values(series):
    num = series.isnull().sum()
    den = len(series)
    return round(num/den, 3)
get_perc_of_missing_values(app_data)
```

```python
# Removing columns where null values are greater than 40%
for col, values in app_data.iteritems():
    if get_perc_of_missing_values(app_data[col]) > 0.40:
        app_data.drop(col, axis=1, inplace=True)
app_data
```

```python
#finding the percentage of null values in each column 
(app_data.isnull().sum() * 100 / len(app_data)).round(2)
```

```python
# Removing External_Source_3 Column as it has almost 20% Missing Values
app_data.drop("EXT_SOURCE_3",axis=1,inplace=True)
```

```python
app_data.describe().T
```

```python
# Since "AMT_GOODS_PRICE"  & "EXT_SOURCE_2" has very low missing values,
# we can use mean values to fill fill those columns 

app_data['AMT_GOODS_PRICE'].fillna((app_data['AMT_GOODS_PRICE'].mean()), inplace=True)
app_data['EXT_SOURCE_2'].fillna((app_data['EXT_SOURCE_2'].mean()), inplace=True)
app_data[["AMT_GOODS_PRICE","EXT_SOURCE_2"]].describe().T
```

```python
# As "AMT_ANNUITY" & "CNT_FAM_MEMBERS" columns have few NaN values, we can fill it with mean value
app_data['AMT_ANNUITY'].fillna((app_data['AMT_ANNUITY'].mean()), inplace=True)
app_data['CNT_FAM_MEMBERS'].fillna((app_data['CNT_FAM_MEMBERS'].mean()), inplace=True)
app_data['DAYS_LAST_PHONE_CHANGE'].fillna((app_data['DAYS_LAST_PHONE_CHANGE'].mean()), inplace=True)
```
